> ### PLOTTING PARIS, FRANCE with STAMEN TERRAIN and ADDING MARKER


```
world_map = folium.Map(location = [48.8566969, 2.3514616], zoom_start = 10, tiles = 'Stamen Terrain')

france = folium.map.FeatureGroup()

france.add_child(folium.CircleMarker([48.8566969, 2.3514616], 
                                             radius = 5, 
                                             color = 'red', 
                                             fill_color = 'Red'))
world_map.add_child(france)
folium.Marker([48.8566969, 2.3514616], popup = 'Paris, France').add_to(world_map)
world_map
```

> ### PLOTTING FRANCE with STAMEN TERRAIN and ADDING POP-UP MARKER


```
# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(COVID.Latitude, COVID.Longitude):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add pop-up text to each marker on the map
latitudes = list(COVID.Latitude)
longitudes = list(COVID.Longitude)
labels = list(COVID.Country)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(world_map)    
    
# add incidents to map
world_map.add_child(incidents)
```

```
from folium import plugins

# let's start again with a clean copy of the map of San Francisco
#sanfran_map = folium.Map(location = [latitude, longitude], zoom_start = 12)

# instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(world_map)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(COVID.Latitude, COVID.Longitude, COVID.Country):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(incidents)

# display map
world_map
```

> ### PLOTTING CHOLOPLETH MAP

```
threshold_scale = np.linspace(COVID['Confirmed'].min(),
                              COVID['Confirmed'].max(),
                              6, dtype=int)
threshold_scale = threshold_scale.tolist() # change the numpy array to a list
threshold_scale[-1] = threshold_scale[-1] + 1 # make sure that the last value of the list is greater than the maximum immigration

# let Folium determine the scale.
world_map = folium.Map(location=[0, 0], zoom_start=2, tiles='Stamen Terrain')
world_map.choropleth(
    geo_data=world_geo,
    data=COVID,
    columns=['Country', 'Confirmed'],
    key_on='feature.properties.name',
    threshold_scale=threshold_scale,
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='COVID Confirm Cases',
    reset=True
)
world_map
```
